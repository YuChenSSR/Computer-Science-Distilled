```python
from sympy import symbols, Sum, simplify

# 定义符号变量
n = symbols('n', integer=True, positive=True)
i = symbols('i', integer=True, positive=True)

# 外层循环操作次数
outer_loop = 2 * (n - 1)

# 内层循环操作次数
inner_loop = Sum(n - i, (i, 1, n - 1))*2

# 总操作次数
total_operations = outer_loop + inner_loop

# 化简结果
simplified_result = simplify(total_operations)

print("外层循环操作次数:", outer_loop)
print("内层循环操作次数:", inner_loop)
print("总操作次数:", total_operations)
print("化简后的总操作次数:", simplified_result)
```

```python
from sympy import symbols, Sum, simplify, latex
from IPython.display import Markdown

# 定义符号变量
n = symbols('n', integer=True, positive=True)
i = symbols('i', integer=True, positive=True)

# 外层循环操作次数
outer_loop = 2 * (n - 1)

# 内层循环操作次数
inner_loop = Sum(n - i, (i, 1, n - 1))*2

# 总操作次数
total_operations = outer_loop + inner_loop

# 化简结果
simplified_result = simplify(total_operations)

# 使用LaTeX格式显示结果
print("外层循环操作次数:")
display(Markdown(f"$${latex(outer_loop)}$$"))

print("内层循环操作次数:")
display(Markdown(f"$${latex(inner_loop)}$$"))

print("总操作次数:")
display(Markdown(f"$${latex(total_operations)}$$"))

print("化简后的总操作次数:")
display(Markdown(f"$${latex(simplified_result)}$$"))
```

### 进一步简化

```python
from sympy import symbols, expand, simplify, latex
from IPython.display import Markdown

# 展开表达式
expanded_expr = expand(simplified_result)

# 化简表达式
simplified_expr = simplify(expanded_expr)

# 使用LaTeX格式显示结果
print("原始表达式:")
display(Markdown(f"$${latex(simplified_result)}$$"))

print("展开后的表达式:")
display(Markdown(f"$${latex(expanded_expr)}$$"))

print("化简后的表达式:")
display(Markdown(f"$${latex(simplified_expr)}$$"))
```

```python
# 将n=8代入化简后的表达式
result = simplified_expr.subs(n, 8)

print("当n=8时,化简后表达式的值:")
display(Markdown(f"$${latex(simplified_expr)}|_{{n=8}} = {result}$$"))


# n =16
result = simplified_expr.subs(n, 16)

print("当n=16时,化简后表达式的值:") 
display(Markdown(f"$${latex(simplified_expr)}|_{{n=16}} = {result}$$"))

# 计算n翻倍对结果倍数的影响
ratio = simplified_expr.subs(n, 16) / simplified_expr.subs(n, 8)

print("当n翻倍时,结果的倍数:")
display(Markdown(f"$${latex(simplified_expr)}|_{{n=16}} / {latex(simplified_expr)}|_{{n=8}} = {ratio}$$"))
```

```python
# 将比例转换为小数
decimal_ratio = float(ratio)

print("结果的倍数(小数):", decimal_ratio)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成n的取值范围
n = np.linspace(1, 100, 10)

# 计算不同时间复杂度函数的值
constant = np.ones_like(n)
logarithmic = np.log(n)
linear = n
linearithmic = n * np.log(n)
quadratic = n ** 2
exponential = 2 ** n

# 绘制时间复杂度曲线图
plt.figure(figsize=(10, 6))
plt.plot(n, constant, label='O(1)')
plt.plot(n, logarithmic, label='O(log n)')
plt.plot(n, linear, label='O(n)')
plt.plot(n, linearithmic, label='O(n log n)')
plt.plot(n, quadratic, label='O(n^2)')
plt.plot(n, exponential, label='O(2^n)')

# 设置图表标题和标签
plt.title('Time Complexity of Common Functions')
plt.xlabel('n')
plt.ylabel('Time Complexity')

# 添加图例
plt.legend()

# 显示图表
plt.show()
```
